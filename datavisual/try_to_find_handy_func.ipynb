## Try to describe the function with its derivative

```python

x_values = []

def get_func(x):
    
    return x**2 +3

def get_nabla(x):
    return 2*x

def get_hessian(x,):
    
    return 2

def calculate_points():
    
    step = 0.001
    x_values = []
    x0 = -1.0
    
    values = []
    
    while x0 < 1:
        
        x= x0 + step
        
        f = get_func(x0) + get_nabla(x)*(x - x0) + 0.5* get_hessian(x) * (x - x0)
        
        x_values.append(x0)
        
        x0 = x
        
        values.append(f)
        
    
    return x_values ,np.array(values)
        
    
    
```

```python
x,y = calculate_points()
```

```python
plt.plot(x,y)
```
